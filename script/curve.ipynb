```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
log_path = '../log.txt'
val_acc = []
with open(log_path, 'r') as f:
    for line in f.readlines():
        if line.startswith('.average accuracy'):
            fields = line.split(':')
            assert len(fields) == 2
            val_acc.append(float(fields[1]))
m = max(val_acc)
idx = [i for i, j in enumerate(val_acc) if j == m][0]

x = np.arange(len(val_acc)) + 1
plt.plot(x, val_acc)
plt.xlabel('epoch')
plt.ylabel('val acc')
plt.title('Validating Accucary')
plt.savefig('val_acc.jpg')
plt.show()
print(' Best accuracy occurs in epoch {} with {}'.format(idx, m))
```
